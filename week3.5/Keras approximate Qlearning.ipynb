```python
env = gym.make("CartPole-v0")
print('Sample state ',env.observation_space.sample())
print('Action types: {0}'.format(env.action_space.n))
```

```python
s = env.reset()
for i in range(500):
    env.render()
    next_state, reward, done, _ = env.step(env.action_space.sample())
    if(done):
        print('Done!')
        break
env.close()
```

```python
agent = Sequential()
agent.add(Dense(32, activation='relu', input_dim=4))
agent.add(Dense(32, activation='relu'))
agent.add(Dense(2, activation='linear'))

agent.compile(loss='mse', optimizer='adam')
```

```python
s = env.reset()
s = s.reshape((1, 4))
q = agent.predict(s)
print(q)
print(np.max(q))
print(np.amax(q))
```
